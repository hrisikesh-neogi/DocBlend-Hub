```python
import os
import sys
from moviepy.editor import VideoFileClip


def convert_video_to_audio_moviepy(video_file, output_ext="wav"):
    """Converts video to audio using MoviePy library
    that uses `ffmpeg` under the hood"""
    filename, ext = os.path.splitext(video_file)
    clip = VideoFileClip(video_file)
    clip.audio.write_audiofile(f"{filename}.{output_ext}")

```

```python
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer, pipeline
import torch

model="openai/whisper-base"
tokenizer = AutoTokenizer.from_pretrained(model)

pipeline = pipeline(
    "automatic-speech-recognition", #task
    model=model,
    tokenizer=tokenizer,
  chunk_length_s=10,
)

prediction = pipeline(
    audio,
    batch_size=8,
    # return_timestamps=True
)["text"]

```

```python
prediction_with_time = pipeline(
    audio,
    batch_size=8,
    return_timestamps=True
)
```

```python
import pandas as pd 

texts = [text["text"] for text in prediction_with_time["chunks"]]
times = [text["timestamp"] for text in prediction_with_time["chunks"]]

data = pd.DataFrame()
data["timestamp"] = times
data["text"] = texts 
```

```python
import re

def extract_start_end_time(time_str):
    # Define a regular expression pattern to match the start and end times
    pattern = r'\((?P<start_time>[\d.]+),\s*(?P<end_time>[\d.]+)\)'

    # Use re.search to find the pattern in the input string
    match = re.search(pattern, time_str)

    if match:
        # Extract start and end times from the named groups in the regex pattern
        start_time = float(match.group('start_time'))
        end_time = float(match.group('end_time'))
        return start_time, end_time
    else:
        # Return None if the pattern is not found in the input string
        return None, None

# # Example usage:
# time_str = "(0.0, 3.72)"
# start_time, end_time = extract_start_end_time(time_str)

# if start_time is not None and end_time is not None:
#     print("Start Time:", start_time)
#     print("End Time:", end_time)
# else:
#     print("Invalid time format or pattern not found.")
```

```python
timestamp = data["timestamp"].astype(str)
time_stamps = timestamp.apply(extract_start_end_time)
start_times = [time[0] for time in time_stamps]
end_times = [time[-1] for time in time_stamps]
```

```python
data["start_time"], data["end_time"] = start_times, end_times
```

```python
data = data.drop("timestamp", axis = 1)
```

```python
import re

def is_youtube_url(input_string):
    # Regular expression pattern for a YouTube video URL
    pattern = r'^(https?://)?(www\.)?(youtube\.com/watch\?v=|youtu\.be/)([a-zA-Z0-9_-]+)(\S*)$'
    
    # Try to match the input string with the pattern
    match = re.match(pattern, input_string)
    
    # If there's a match and the video ID is found, return True
    if match and match.group(4):
        return True
    else:
        return False

# Example usage:
url = "https://youtu.be/5p248yoa3oE?si=s0uCo9nxcjF3lOlK"
if is_youtube_url(url):
    print(f"'{url}' is a valid YouTube video URL.")
else:
    print(f"'{url}' is not a valid YouTube video URL.")
```

```python
from urllib.parse import urlparse

def is_youtube_url(url):
    # Parse the URL
    parsed_url = urlparse(url)

    # Check if the domain is youtube.com or youtu.be
    if parsed_url.netloc.endswith("youtube.com") or parsed_url.netloc == "youtu.be":
        return True
    else:
        return False

# Example usage:
url = "https://youtu.be/5p248yoa3oE?si=s0uCo9nxcjF3lOlK"
if is_youtube_url(url):
    print(f"'{url}' is a YouTube URL.")
else:
    print(f"'{url}' is not a YouTube URL.")
```
